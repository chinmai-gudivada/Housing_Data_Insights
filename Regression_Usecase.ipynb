```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import numpy as np
```

```python
linear_model = LinearRegression()

linear_model.fit(X_train, y_train)
```

```python
linear_pred = linear_model.predict(X_test)
```

```python
linear_mse = mean_squared_error(y_test, linear_pred)
linear_rmse = np.sqrt(linear_mse)
linear_mae = mean_absolute_error(y_test, linear_pred)
linear_mape = mean_absolute_percentage_error(y_test, linear_pred)
linear_r2 = r2_score(y_test, linear_pred)
```

```python
poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(X_train)
x_test_poly = poly.transform(X_test)
```

```python
x_poly = LinearRegression()

x_poly.fit(x_train_poly, y_train)
```

```python
poly_pred = x_poly.predict(x_test_poly)
```

```python
poly_mse = mean_squared_error(y_test, poly_pred)
poly_rmse = np.sqrt(poly_mse)
poly_mae = mean_absolute_error(y_test, poly_pred)
poly_mape = mean_absolute_percentage_error(y_test, poly_pred)
poly_r2 = r2_score(y_test, poly_pred)
```

```python
lasso_model = Lasso()

lasso_model.fit(X_train, y_train)
```

```python
lasso_pred = lasso_model.predict(X_test)
```

```python
lasso_mse = mean_squared_error(y_test, lasso_pred)
lasso_rmse = np.sqrt(lasso_mse)
lasso_mae = mean_absolute_error(y_test, lasso_pred)
lasso_mape = mean_absolute_percentage_error(y_test, lasso_pred)
lasso_r2 = r2_score(y_test, lasso_pred)
```

```python
ridge_model = Ridge()

ridge_model.fit(X_train, y_train)
```

```python
ridge_pred = ridge_model.predict(X_test)
```

```python
ridge_mse = mean_squared_error(y_test, ridge_pred)
ridge_rmse = np.sqrt(ridge_mse)
ridge_mae = mean_absolute_error(y_test, ridge_pred)
ridge_mape = mean_absolute_percentage_error(y_test, ridge_pred)
ridge_r2 = r2_score(y_test, ridge_pred)
```

## Outliers

```python
df = data.copy()
```

```python
df.info()
```

```python
def detect_outliers(df, threshold=1.5):

  outliers_dict = {}

  for column in df.select_dtypes(include=['int64', 'float64']).columns:

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR

    column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    outliers_dict[column] = column_outliers

  return outliers_dict
```

```python
outliers_dict = detect_outliers(df)
```

```python
total_entries = len(df)

outlier_summary = {}

for column, outlier in outliers_dict.items():

  num_outliers = len(outlier)
  outlier_percentage = (num_outliers / total_entries) * 100

  outlier_summary[column] = [num_outliers, outlier_percentage]
```

```python
all_outliers = pd.concat(outliers_dict.values()).drop_duplicates()
num_total_outliers = len(all_outliers)
total_outlier_percentage = (num_total_outliers / total_entries) * 100
```

```python
print('Outlier Summary for each Column:')
for column, summary in outlier_summary.items():
  print(f'{column}: Count = {summary[0]}, Percentage = {summary[1]:.2f}%')

print(f'\nTotal Outliers: Count = {num_total_outliers}, Percentage = {total_outlier_percentage:.2f}%')
```

```python
df_cleaned = df.drop(all_outliers.index)
```
